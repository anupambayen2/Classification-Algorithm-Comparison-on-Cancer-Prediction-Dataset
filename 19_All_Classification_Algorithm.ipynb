## All Algorithm Testing

### Importing the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Importing the dataset

```python
dataset = pd.read_csv('dataset/cancer_data.csv')
```

```python
X = dataset.iloc[:,1:10].values
y = dataset.iloc[:,-1].values
```

### Splitting the dataset inot train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression(random_state=0)
logistic.fit(X_train,y_train)
```

```python
log_pred = logistic.predict(X_test)
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
decision = DecisionTreeClassifier(random_state=0)
```

```python
decision.fit(X_train,y_train)
```

```python
dec_pred = decision.predict(X_test)
```

### Random Forest

```python
from sklearn.svm import SVC
sv_c = SVC(kernel='linear',random_state=0)
```

```python
sv_c.fit(X_train,y_train)
```

```python
svc_pred = sv_c.predict(X_test)
```

### Kernel SVM

```python
from sklearn.svm import SVC
svc_kernel = SVC(kernel='rbf',random_state=0)
svc_kernel.fit(X_train,y_train)
```

```python
kernel_pred = svc_kernel.predict(X_test)
```

### K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
neighbor = KNeighborsClassifier()
```

```python
neighbor.fit(X_train,y_train)
```

```python
neighbor_pred = neighbor.predict(X_test)
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
```

```python
nb.fit(X_train,y_train)
```

```python
nb_pred = nb.predict(X_test)
```

### Measuring the Performance 

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

#### Logistic Regression

```python
accuracy_score(log_pred,y_test)
```

#### Decision Tree

```python
accuracy_score(dec_pred,y_test)
```

#### Random Forest

```python
accuracy_score(svc_pred,y_test)
```

#### Kernel SVM

```python
accuracy_score(kernel_pred,y_test)
```

#### K-Nearest Neighbors

```python
accuracy_score(neighbor_pred,y_test)
```

#### Naive Bayes

```python
accuracy_score(nb_pred,y_test)
```
